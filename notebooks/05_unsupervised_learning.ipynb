```python
import pandas as pd
data=pd.read_csv('../data/processed_data.csv')

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
```

```python
X=data.drop('target',axis=1)
y=data['target']
```

Apply K-Means Clustering (elbow method to determine K)

```python
wcss=[]
K_range=range(1,11)
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

plt.figure(figsize=(6,4))
plt.plot(K_range, wcss, marker='o')
plt.xlabel("K")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

Perform Hierarchical Clustering (dendrogram analysis).


```python
hc_model = AgglomerativeClustering(n_clusters=3, linkage='ward')
hc_labels = hc_model.fit_predict(X)

plt.figure(figsize=(8, 4))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrogram")
plt.show()
```

Compare clusters with actual disease labels.

```python
kmeans_score = adjusted_rand_score(y,kmeans_labels)
print("K-Means:",kmeans_score)

hc_score = adjusted_rand_score(y,hc_labels)
print("Hierarchical:",hc_score)
```
